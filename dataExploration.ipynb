```python
import glob
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
datasetRoot = 'dataset/liver/'
setName = 'train'
patientNo = 0
sliceNo = 10

patientSliceImagePath = datasetRoot+'/'+setName+'/'+'liver_'+str(patientNo) + '/'
patientSliceLabelPath = datasetRoot+'/'+'labels'+'/'+'liver_'+str(patientNo) + '/'
sliceFilePath = sorted(glob.glob(patientSliceImagePath+'/*.png'))

print('Total slice images:', len(sliceFilePath),)
```

```python
sliceImagePath = patientSliceImagePath + str(sliceNo).zfill(3) + '.png'
sliceImage = cv2.imread(sliceImagePath)
sliceLabelPath = patientSliceLabelPath + str(sliceNo).zfill(3) + '.png'
sliceLabels = cv2.imread(sliceLabelPath, cv2.IMREAD_GRAYSCALE)

plt.figure()
plt.imshow(sliceImage)

plt.figure()
plt.imshow(sliceLabels, vmin=0, vmax=255,cmap='gray')
```
